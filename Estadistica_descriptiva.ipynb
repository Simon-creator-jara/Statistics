Escalamiento dde datos numricos

Transformaciones lineales

```python
import timeit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model

X,y = datasets.load_diabetes(return_X_y=True)
raw=X[:,None,2]
```

```python
#Reglas de escalamiento
max_raw = max(raw)
min_raw = min(raw)
scaled = (2*raw - max_raw - min_raw)/(max_raw - min_raw)

fig, axs = plt.subplots(2,1,sharex=True)
axs[0].hist(raw)
axs[1].hist(scaled)
```

```python
#Modelos para entrenamiento

def train_raw():
  linear_model.LinearRegression().fit(raw,y)

def train_scaled():
  linear_model.LinearRegression().fit(scaled,y)
```

```python
raw_time = timeit.timeit(train_raw,number=100)
scaled_time = timeit.timeit(train_scaled, number=100)
print('train raw: {}'.format(raw_time))
print('train scaled: {}'.format(scaled_time))
```

max - min: mejor para datos uniformemente distribuidos

Z-score = mejor para datos distribuidos "normalmente" (forma de campana de gauss)

# Covarianza

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler ## Para normalizacin de datos

iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris, hue='species')
```

```python
iris.columns
```

```python
scaler=StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
)
scaled.T
```

```python
matriz_covarianza=np.cov(scaled.T)
matriz_covarianza
```

Matriz de covarianza

```python
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(matriz_covarianza,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt ='.2f',
                 annot_kws={'size':12},
                 yticklabels=['sepal_lenght','sepal_widht','petal_lenght','petal_widht'],
                 xticklabels=['sepal_lenght','sepal_widht','petal_lenght','petal_widht']
                 )
```

# PCA

```python
sns.jointplot(x=iris['petal_length'], y =iris['petal_width'])
sns.jointplot(x = scaled[:,2], y=scaled[:,3])
```

```python
eigen_values, eigen_vectors = np.linalg.eig(matriz_covarianza)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
variance_explained=[]
for i in eigen_values:
  variance_explained.append((i/sum(eigen_values))*100)

print(variance_explained)
```

```python
from sklearn.decomposition import PCA


pca=PCA(n_components=2)
pca.fit(scaled)
```

```python
pca.explained_variance_ratio_
```

```python
reduced_scaled = pca.transform(scaled)
```

```python
reduced_scaled
```
